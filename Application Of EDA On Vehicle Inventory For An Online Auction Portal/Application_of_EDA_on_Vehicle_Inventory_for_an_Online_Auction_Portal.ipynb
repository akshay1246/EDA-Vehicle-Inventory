Joining the State Frequency Data with Abbreviations

```python
state=df.groupby(['state']).size().reset_index().rename(columns={0:'counts'})
state #we shall now join the state wise frequency count along with their abbreviation
```

```python
df1
```

```python
statemap=df1.merge(state,right_on='state',left_on='Region')
```

```python
statemap #we can remove one column 
```

```python
statemap=statemap.drop(['state'], axis=1)
```

```python
statemap.head()
```

# Plotting the State Frequency on the USA Map

```python
fig = px.choropleth(statemap,
                    locations = statemap['Abb'], 
                    locationmode = "USA-states", 
                    scope = "usa",
                    color = statemap['counts'],
                    color_continuous_scale = "RdYlGn_r",  
                    )

#Updating the title and font
fig.update_layout(
      title_text = 'Overall Vehicle Distribution over US',
      title_font_family="Times New Roman",
      title_font_size = 22,
      title_font_color="black", 
      title_x=0.45, 
         )

#Displaying the USA plot
fig.show()
```

# **Joining the Main Data with the Abbreviations**

```python
#Left outer join
finaldf=df.merge(df1, how='left', left_on='state',right_on='Region')
```

```python
finaldf.head()
```

```python
finaldf=finaldf.drop(['state'],axis=1) #dropping state column
```

## Filtering the Joined Dataset with Brand as FORD

```python
ford=finaldf[finaldf['brand']=='ford']
```

```python
ford.head()
```

```python
fy=fy.sort_values(by=['year'], ascending=False)
fy=fy.head(10)
```

```python
fig=px.bar(fy, x='year',y='counts', title='Year wise distribution of Ford cars')
fig.show()
```

```python
fprice=df.groupby(['title_status'])['price'].mean().reset_index().rename(columns={0:'counts'})	
fprice
```
